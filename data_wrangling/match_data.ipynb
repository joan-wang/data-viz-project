## Try to match country names from DOL data (Sweat & Toil) to World Bank data (indicators)
Use DOL countries table as the master list

```python
net_migration = pd.read_csv('data/API_SM.POP.NETM_DS2_en_csv_v2.csv', header=2)
countries = pd.read_csv('data/Countries.csv', index_col=0)
```

C and S appear to be typos in DOL data and are removed.

```python
countries = countries[(countries.name !='C') & (countries.name !='S')]
```

```python
net_migration.head() # most recent data is 2012
countries.head()
```

Join datasets to see which countries in DOL data do not match with World Bank data

```python
joined =countries.merge(net_migration[['Country Name', 'Indicator Name', '2012']], how='left', left_on='name', right_on='Country Name')
joined.shape
```

```python
mismatch = joined[pd.isnull(joined['2012'])]
mismatch.shape
mismatch
```

Create name_WB column in countries table, containing names that match World Bank data.   

Manually matching those that were unmatched. Some countries are not represented in World Bank data or have missing values in specified dataset.

```python
countries.shape
countries['name_WB'] = countries['name'] 
countries.loc[countries.name == 'Burma','name_WB'] = 'Myanmar'
countries.loc[countries.name == 'North Korea','name_WB'] = "Korea, Dem. Peopleâ€™s Rep."
countries.loc[countries.name == 'Macedonia','name_WB'] = 'Macedonia, FYR'
countries.loc[countries.name == 'Moldova, Republic of','name_WB'] = 'Moldova'
countries.loc[countries.name == 'Russia','name_WB'] = 'Russian Federation'
countries.loc[countries.name == 'Egypt','name_WB'] = 'Egypt, Arab Rep.'
countries.loc[countries.name == 'Iran','name_WB'] = 'Iran, Islamic Rep.'
countries.loc[countries.name == 'West Bank and the Gaza Strip','name_WB'] = 'West Bank and Gaza'
countries.loc[countries.name == 'Yemen','name_WB'] = 'Yemen, Rep.'
countries.loc[countries.name == 'Congo, Democratic Republic of the','name_WB'] = 'Congo, Dem. Rep.'
countries.loc[countries.name == 'Congo, Republic of the','name_WB'] = 'Congo, Rep.'
countries.loc[countries.name == 'Gambia','name_WB'] = 'Gambia, The'
countries.loc[countries.name == 'Saint Lucia','name_WB'] = 'St. Lucia'
countries.loc[countries.name == 'Saint Vincent and the Grenadines','name_WB'] = 'St. Vincent and the Grenadines'
countries.loc[countries.name == 'Venezuela','name_WB'] = 'Venezuela, RB'

countries.shape
```

Join again to see if we get expected results.

```python
joined2 = countries.merge(net_migration[['Country Name', 'Indicator Name', '2012']], how='inner', left_on='name_WB', right_on='Country Name')
joined2.shape
mismatch2 = joined2[pd.isnull(joined2['2012'])]
mismatch2.shape
mismatch2
```
